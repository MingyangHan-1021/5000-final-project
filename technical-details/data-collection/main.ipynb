# Code 

The code used in this section is adapted from a research[@Abodayeh2023WebSF] and has been modified to focus specifically on Lionel Messi’s club and international data. The script retrieves and parses HTML content from Wikipedia pages using the requests and BeautifulSoup libraries, and then uses pandas.read_html() to convert the tables into structured DataFrames, providing a foundation for subsequent analysis.


## Data of Lionel Messi's Club Career or international Career

```python
# import necessary libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python
# URL of the Wikipedia page 
url = "https://en.wikipedia.org/wiki/Lionel_Messi"

# HTTP headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

# Send the GET request
response = requests.get(url, headers=headers)
response.raise_for_status()

# Parse the HTML content
soup = BeautifulSoup(response.content, "lxml")

# Find all tables with class "wikitable"
tables = soup.find_all("table", {"class": "wikitable"})

# Locate the "Club career statistics" table
club_table = None
for table in tables:
    headers = [th.get_text(strip=True) for th in table.find_all("th")]
    # Club career table usually has columns like "Season", "Club", "League", "Apps", "Goals"
    if any("Season" in h or "Club" in h for h in headers):
        club_table = table
        break

# Raise an exception if the table was not found
if club_table is None:
    raise Exception("Club career table not found")

# Use pandas to read the HTML table into a DataFrame
df = pd.read_html(str(club_table))[0]

# Display column names
print(df.columns)

# Save the DataFrame as a CSV file
# df.to_csv("/Users/hanmingyang/Desktop/lionel_messi_club_career.csv", index=False, encoding="utf-8-sig")
# print("CSV file has been saved as lionel_messi_club_career.csv")
```

## Data of Lionel Messi's Player profile

```python

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find the "Player profile" section header
player_profile_header = soup.find(id="Player_profile")
if not player_profile_header:
    raise ValueError("Player profile section not found")

# Initialize data dictionary
data = {
    "section": "Player profile",
    "paragraphs": []
}

# Iterate over the sibling elements after the header, capturing only the first 7 paragraphs
paragraph_count = 0
for sibling in player_profile_header.parent.find_next_siblings():
    if sibling.name == "h2":
        break
    if sibling.name == "p":
        text = sibling.get_text(strip=True)
        if text:
            data["paragraphs"].append(text)
            paragraph_count += 1
            if paragraph_count >= 7:  # Limit the number of paragraphs
                break

# Save the data as a JSON file
# with open("/Users/hanmingyang/Desktop/Lionel_Messi_Player_Profile.json", "w", encoding="utf-8") as f:
    # json.dump(data, f, ensure_ascii=False, indent=4)

# print("Saved as Lionel_Messi_Player_Profile.json")
```
